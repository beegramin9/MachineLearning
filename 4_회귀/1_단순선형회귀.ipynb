# 단순 선형 회귀
- 아버지와 아들의 키 - Galton

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('http://www.randomservices.org/random/data/Galton.txt', sep='\t')
df.head()
```

```python
df.info()
```

```python
df = df[df.Gender == 'M']
height = df[['Father','Height']].rename(columns={'Height':'Son'})
height.Father = height.Father * 2.54
height.Son = height.Son * 2.54
height.describe()
```

```python
plt.scatter(height.Father, height.Son)
plt.xlabel('Father\'s height')
plt.ylabel('Son\'s height')
plt.grid()
plt.show()
```

### 고전적 방법: Numpy를 이용해서 회귀선 구하기

```python
# Numpy 최소자승법: 오차 제곱의 합을 최소화
A = np.vstack([height.Father, np.ones(len(height.Father))]).T
A.shape
```

```python
reg = np.linalg.lstsq(A, height.Son , rcond=None)
reg
```

```python
# 기울기와 절편
m, c = reg[0]
```

```python
# 잔차 제곱의 합 (Residual Sum of Square)
rss = reg[1][0]
```

```python
print(f'기울기: {m:.4f}, 절편: {c:.4f}, 잔차제곱: {rss:.4f}')
```

```python
plt.figure(figsize=(8,7))
plt.scatter(height.Father, height.Son)
# x좌표, y좌표. 점 2개만 알면 직선 그릴 수 있음
plt.plot([150,200], [m*150+c, m*200+c], 'r', lw=3)
plt.title('The height of Father and Son')
plt.xlabel('Father\'s height')
plt.ylabel('Son\'s height')
plt.grid()
plt.show()
```

### sklearn으로 구하기

```python
from sklearn.linear_model import LinearRegression
```

```python
X = height.Father.values.reshape(-1,1)
""" X가 1차원이면 독립변수로 줄 수 없음. 항상 2차원으로 만들어줘야 함 """
y = height.Son.values # 1차원
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
# 해당 클래스의 생성자, 속성, 메소드 알아내는 함수
""" m, c, rss를 찾아야 함 """
dir(lr)
```

y = WX + b (W: weight, b: bias) <br>
X 가 m x n 이면 W는 n x m

```python
weight, bias, rss = lr.coef_, lr.intercept_, lr._residues
# weight는 W 행렬이니까 Array로 나온다
print(f'기울기: {weight[0]:.4f}, 절편: {bias:.4f}, 잔차제곱: {rss:.4f}')
```

### R Squared(R2 score): 선형식이 얼마나 잘 대표하는지 척도

```python
score = lr.score(X,y)
# 0.1531, 별로 좋지 않음
print(f'R_squared score: {score:.4f}')
```

### 단순선형회귀 - 당뇨병 데이터셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
df
```

```python
diabetes.target[:5]
```

```python
# BMI 수치만 가지고 회귀 (단순선형회귀)
X = df.bmi.to_numpy().reshape(-1,1)
# 2차행렬로 만들어줘야 함
print(X.shape)
y = diabetes.target
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
weight, bias, rss = lr.coef_, lr.intercept_, lr._residues
# weight는 W 행렬이니까 Array로 나온다
print(f'기울기: {weight[0]:.4f}, 절편: {bias:.4f}, 잔차제곱: {rss:.4f}')
```

### 훈련/테스트 데이터 셋 분리후 학습/에측/평가

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size= 0.05, random_state = 2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print(mse, rmse)
```

```python
score = r2_score(y_test, pred)
print(score)
```

### 그래프로 확인

```python
weight, bias, rss = lr.coef_, lr.intercept_, lr._residues
# weight는 W 행렬이니까 Array로 나온다
print(f'기울기: {weight[0]:.4f}, 절편: {bias:.4f}, 잔차제곱: {rss:.4f}')
```

```python
plt.scatter(X_train, y_train)
plt.plot([-0.1,0.2],[-0.1*weight + bias, 0.2*weight + bias], 'r', lw=3)
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.1,0.2],[-0.1*weight + bias, 0.2*weight + bias], 'r', lw=3)
plt.grid()
plt.show()
```
